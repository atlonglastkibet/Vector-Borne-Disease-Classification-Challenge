```python
import plotly.graph_objects as go
import pandas as pd

# Define the nodes
labels = ['A', 'B', 'C', 'D', 'E', 'F']

# Define the flows (source node, target node, value)
source = [0, 0, 1, 2, 2, 3]
target = [1, 2, 3, 3, 4, 5]
value = [8, 15, 10, 6, 9, 5]

# Colors can be specified for each link
link_colors = ['rgba(31, 119, 180, 0.4)', 'rgba(255, 127, 14, 0.4)', 
               'rgba(44, 160, 44, 0.4)', 'rgba(214, 39, 40, 0.4)',
               'rgba(148, 103, 189, 0.4)', 'rgba(140, 86, 75, 0.4)']

# Create the sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color="blue"
    ),
    link=dict(
        source=source,
        target=target,
        value=value,
        color=link_colors
    )
)])

# Update the layout
fig.update_layout(
    title_text="Basic Sankey Diagram",
    font_size=12,
    height=600,
    width=800
)

# Save to HTML file (will work in any environment)
fig.write_html("sankey_diagram.html")
print("Sankey diagram saved to 'sankey_diagram.html'. Open this file in your browser to view it.")

# If you're in a Jupyter notebook and still want to try displaying it inline:
try:
    fig.show()
except ValueError:
    print("Couldn't display the figure inline. Please open the saved HTML file instead.")
```
