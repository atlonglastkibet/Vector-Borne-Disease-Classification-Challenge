```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
cols = df.columns[df.columns != 'prognosis']
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
for i, col in enumerate(cols):
    ax = plt.subplot(8,10 , i + 1)
    ax.set_title(col)
    ax.set_xlabel('Count')
    ax.set_ylabel('Value')
    sns.kdeplot(df[col], ax=ax, color='blue', alpha=0.5)
    plt.tight_layout()
```
